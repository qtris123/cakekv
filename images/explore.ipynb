```python
import re

def tokenize(text):
    """Simple whitespace + punctuation tokenizer."""
    return re.findall(r"\w+", text.lower())

def f1_score(pred, ref):
    pred_tokens = tokenize(pred)
    ref_tokens = tokenize(ref)
    
    pred_set = set(pred_tokens)
    ref_set = set(ref_tokens)
    
    overlap = len(pred_set & ref_set)
    if overlap == 0:
        return 0.0, 0.0, 0.0

    precision = overlap / len(pred_set)
    recall = overlap / len(ref_set)
    f1 = 2 * precision * recall / (precision + recall)
    return f1, precision, recall

# Example
for i in range(len(lines)):
    pred = lines[i]["output"]
    ref  = refs[i]["answers"][0]
    f1, p, r = f1_score(pred, ref)
    print(f"Precision={p:.2f}, Recall={r:.2f}, F1={f1:.2f}")
```

```python
import numpy as np

# Holds lists of rows for each layer
decode_rows_per_layer = {}  # layer_idx -> list of np.array rows [H,S]

def _pad_to_width(arr: np.ndarray, width: int) -> np.ndarray:
    """Pad a row [H,S] to [H,width] with zeros on the right."""
    H,_, S = arr.shape
    if S == width:
        return arr
    out = np.zeros((H,1, width), dtype=arr.dtype)
    out[:,:, :S] = arr
    return out

rows = [item["tensor"] for item in atttn0[0][1:]]
rows = [_pad_to_width(r, rows[-1].shape[-1]) for r in rows]
square = np.stack(rows, axis = 0)
square = square.squeeze(axis = 2)
```
